Pyplot, PyCall import of Networkx:

```julia
using PyCall
using PyPlot

PyCall.@pyimport networkx as nx

PyPlot.plt.clf()
DG=nx.DiGraph()
DG[:add_weighted_edges_from]([(1,2,0.5), (3,1,0.75)])
nx.draw(DG)
PyPlot.plt.gcf()
```

Pyplot:

```julia
PyPlot.plt.clf()
x = PyPlot.linspace(0, 2*pi)
PyPlot.plot(x, sin(x))
PyPlot.plot(x, cos(x), "ro")
PyPlot.title("Two familiar functions")
#legend();
PyPlot.plt.gcf()
```
